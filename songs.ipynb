МАТРИЦА КОРРЕЛЯЦИИ 
Строим матрицу корреляции только с числовыми колонками. Которые мы вывели выше.

```python
# Выбор только числовых колонок для корреляции
numeric_columns = ['popularity', 'duration_min', 'danceability', 'energy', 'key', 'loudness', 
                   'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo']

# Создание подмножества данных
numeric_data = data[numeric_columns]

# Вычисление корреляции
correlation_matrix = numeric_data.corr()

# Построение тепловой карты
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)

# Добавление заголовка
plt.title("Матрица корреляции", fontsize=16)
plt.show()
```

ВЫВОДЫ АНАЛИЗА МАТРИЦЫ КОРРЕЛЯЦИИ. 
1)Сильно положительные корреляции (значения близки к +1):

1.1)energy и loudness:
Коэффициент корреляции показывает сильную связь между энергией и громкостью трека. Это интуитивно понятно, поскольку более энергичные треки обычно звучат громче.
Подходит для анализа "зажигательных" песен.
1.2)danceability и valence:
Треки с высоким уровнем танцевальности часто ассоциируются с положительным настроением. Это важно при создании плейлистов для вечеринок или поднятия настроения.
2)Сильно отрицательные корреляции (значения близки к -1):

2.1)acousticness и energy:
Чем выше акустичность трека, тем меньше энергии. Акустические треки часто звучат спокойнее и имеют более мягкую энергетику.
Это полезно при создании расслабляющих плейлистов или анализа спокойной музыки.
2.2)acousticness и loudness:
Треки с высокой акустичностью обычно менее громкие, что подтверждает их мягкость.
2.3)instrumentalness и valence:
Более инструментальные треки имеют тенденцию быть менее позитивными. Это может быть связано с отсутствием вокала, что влияет на восприятие настроения.
3)Умеренные положительные корреляции (значения около +0.5):

3.1)popularity и danceability:
Танцевальные треки часто становятся более популярными. Это может быть полезно для прогнозирования популярности трека на основе его характеристик.
3.2)popularity и valence:
Позитивные треки имеют тенденцию быть более популярными. Это может объясняться предпочтением слушателей к более жизнерадостной музыке.
3.3)tempo и energy:
Более быстрый темп часто сопровождается увеличением энергии трека. Такие треки воспринимаются как более "зажигательные".
4)Умеренные отрицательные корреляции (значения около -0.5):
4.1)acousticness и popularity:
Высокая акустичность может снижать популярность, так как современные популярные треки часто включают электронные элементы.
4.2)danceability и liveness:
"Живость" трека (запись на концерте или вживую) может снижать танцевальность. Это связано с акустической средой и меньшей обработкой.
5)Слабые или отсутствующие корреляции (значения около 0):
5.1)key с другими параметрами:
Мажор или минор (ключ трека) практически не влияет на энергетику, танцевальность или популярность. Это подтверждает, что слушатели воспринимают музыку более комплексно.
5.2)speechiness и valence:
Наличие речи в треке не сильно влияет на его настроение.

Гипотеза 1: Длина трека влияет на его популярность.

Треки слишком большой длины могут утомлять слушателей и быть менее популярными.
Очень короткие треки могут не успевать создать эмоциональный отклик.
Возможно, наиболее популярными являются треки со средней длиной, например, 3–4 минуты. 

```python
# Построение графика зависимости популярности от длины трека
plt.figure(figsize=(10, 6))
sns.scatterplot(x='duration_min', y='popularity', data=data, alpha=0.5)
plt.title('Зависимость популярности от длины трека', fontsize=14)
plt.xlabel('Длина трека (мин)')
plt.ylabel('Популярность')
plt.show()
```

Проверим среднюю популярность для каждого интервала.
Разделим треки по интервалам длины:

Короткие треки: до 2 минут.
Средние треки: от 2 до 5 минут.
Длинные треки: более 5 минут.

Строим столбчатую диаграмму

```python
# Создание категорий по длине треков
data['length_category'] = pd.cut(data['duration_min'], 
                                 bins=[0, 120000, 300000, data['duration_min'].max()], 
                                 labels=['Короткие', 'Средние', 'Длинные'])

# Вычисление средней популярности для каждой категории
length_popularity = data.groupby('length_category')['popularity'].mean()

# Построение столбчатой диаграммы
plt.figure(figsize=(8, 5))
length_popularity.plot(kind='bar', color=['blue', 'green', 'red'], alpha=0.7)
plt.title('Средняя популярность по длине трека', fontsize=14)
plt.xlabel('Категория длины трека')
plt.ylabel('Средняя популярность')
plt.xticks(rotation=0)
plt.show()
```

ВЫВОД: НАИБОЛЕЕ ПОПУЛЯРНЫ ПЕСНИ СРЕДНЕЙ(2-5 МИНУТ) ДЛИНЫ. А песни больше 5 минут и более наименее популярны. 

Гипотеза 2: Популярность песни зависит от её жанра

```python
# Подсчёт количества уникальных жанров
unique_genres = data['track_genre'].nunique()
print(f"Общее количество жанров: {unique_genres}")
```

Для того чтобы определить топ-10 популярных жанров, сгруппируем данные по колонке track_genre и посчитаем среднюю популярность для каждого жанра.

Для определения 10 наименее популярных жанров мы сгруппируем данные по колонке track_genre и рассчитаем среднюю популярность каждого жанра.

```python
# Группировка данных по жанрам и подсчёт средней популярности
genre_popularity = data.groupby('track_genre')['popularity'].mean().sort_values()

# Топ-10 наименее популярных жанров
bottom_10_genres = genre_popularity.head(10)

# Вывод результатов
print("Топ-10 наименее популярных жанров:")
print(bottom_10_genres)

# Построение графика
plt.figure(figsize=(10, 6))
bottom_10_genres.plot(kind='bar', color='lightcoral', alpha=0.8)
plt.title('Топ-10 наименее популярных жанров', fontsize=16)
plt.xlabel('Жанры')
plt.ylabel('Средняя популярность')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Импортирую библиотеки:

Keras — это высокоуровневая библиотека для построения нейронных сетей на основе TensorFlow.
Модель создается через объект keras.Sequential, в который последовательно добавляются слои. 

Первый слой: 

Полносвязный слой с 64 нейронами.
Применяет функцию активации ReLU (Rectified Linear Unit), которая улучшает способность сети обучаться, обнуляя отрицательные значения.
input_shape=(X_train.shape[1],): Указывает размерность входных данных. X_train.shape[1] — количество признаков в данных.

Второй слой:

Еще один полносвязный слой с 64 нейронами и функцией активации ReLU.
Добавление второго слоя помогает модели лучше улавливать сложные зависимости.

Выходной слой:

Полносвязный слой с 1 нейроном.
используется для задачи регрессии, где нужно предсказать численное значение.

```python
import tensorflow as tf
from tensorflow import keras

# Импортируем необходимые классы и функции
from sklearn.preprocessing import StandardScaler

# Создаем объект стандартизации данных
scaler = StandardScaler()

# Применяем стандартизацию к тренировочным данным
X_train = scaler.fit_transform(X_train)

# Применяем стандартизацию к тестовым данным
X_test = scaler.transform(X_test)

# Создаем модель Keras
model = keras.Sequential([
    # Первый слой
    keras.layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
    # Второй слой
    keras.layers.Dense(64, activation='relu'),
    # Третий слой
    keras.layers.Dense(1),
])

# Компилируем модель с оптимизатором Adam и метрикой потерь Mean Squared Error
model.compile(optimizer='adam', loss='mse')
```

```python
# обучение модели
history = model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)

# потери на тестовых данных
loss_test = model.evaluate(X_test, y_test)
```

Permutation Importance

Для анализа важности признаков в нейронной сети можно использовать метод вычисления изменения потерь (loss) при изменении значения конкретного признака. Один из подходов заключается в использовании функции под названием Permutation Importance. Этот метод проверяет, насколько влияет перестановка значений конкретного признака на результат модели. Это можно реализовать следующим образом:
Базовая оценка (baseline_loss):
Вычисляется потеря модели без изменений данных.
Перестановка значений:
Значения каждого признака в тестовом наборе случайно перемешиваются. Это нарушает связь между этим признаком и целевой переменной.
Оценка изменений:
Разница между потерями модели на исходных и измененных данных показывает влияние данного признака на производительность.
Сортировка:
Результаты сортируются по величине изменений, чтобы выявить наиболее значимые признаки.

```python
# Функция для расчета важности признаков
def compute_feature_importance(model, X_test, y_test, feature_names):
    """
    Вычисление важности признаков путем изменения значений одного признака
    и оценки влияния на потери.
    """
    baseline_loss = model.evaluate(X_test, y_test, verbose=0)  # Базовые потери
    feature_importances = []

    for i in range(X_test.shape[1]):
        X_test_permuted = X_test.copy()
        np.random.shuffle(X_test_permuted[:, i])  # Перестановка значений конкретного признака
        permuted_loss = model.evaluate(X_test_permuted, y_test, verbose=0)
        importance = permuted_loss - baseline_loss  # Изменение потерь
        feature_importances.append(importance)

    # Возвращаем важность и имена признаков
    return dict(zip(feature_names, feature_importances))

# Список имен признаков
feature_names = X.columns.tolist()  # Если X - DataFrame, иначе используйте список имен

# Вычисление важности
feature_importances = compute_feature_importance(model, X_test, y_test, feature_names)

# Сортируем и выводим важные признаки
sorted_importances = sorted(feature_importances.items(), key=lambda x: x[1], reverse=True)
print("Важность признаков:")
for feature, importance in sorted_importances:
    print(f"{feature}: {importance:.4f}")
```

```python
# Преобразование в удобный формат для построения
features = [item[0] for item in sorted_importances]
importances = [item[1] for item in sorted_importances]

# Построение графика
plt.figure(figsize=(12, 6))
plt.barh(features, importances, color='skyblue')
plt.xlabel('Влияние на потери (Loss)')
plt.ylabel('Признаки')
plt.title('Важность признаков')
plt.gca().invert_yaxis()  # Инвертируем ось для лучшего восприятия
plt.show()
```

После обучения модели Random Forest Regressor, можно использовать следующие метрики для оценки её качества:
Mean Squared Error (MSE) — средний квадрат ошибки
Root Mean Squared Error (RMSE) — корень из среднего квадрата ошибки
Mean Absolute Error (MAE) — средняя абсолютная ошибка
R² (коэффициент детерминации) — доля объяснённой дисперсии

Интерпретация метрик:
MSE: Чем меньше, тем лучше. Хорошо показывает большие отклонения, так как квадратичные ошибки больше штрафуются.
RMSE: Корень из MSE — значение в тех же единицах, что и целевая переменная.
MAE: Среднее отклонение предсказаний от истинных значений. Более устойчив к выбросам.
R²: Значение в диапазоне [0,1], где 1 означает идеальное предсказание. Может быть отрицательным, если модель хуже среднего.

Мы создаем модель GradientBoostingRegressor с параметрами:

n_estimators=1000: количество деревьев в ансамбле.
learning_rate=0.1: скорость обучения, регулирующая вклад каждого дерева.
max_depth=5: максимальная глубина каждого дерева.
min_samples_split=5: минимальное количество образцов, необходимых для разделения узла.
min_samples_leaf=2: минимальное количество образцов в листовом узле.
random_state=0: фиксирует случайность для воспроизводимости.

mae_scorer: пользовательская метрика для MAE, созданная с помощью make_scorer. Это позволяет использовать её в функции cross_validate.

cv=5: 5-кратная кросс-валидация (данные делятся на 5 частей: 4 используются для обучения, 1 для тестирования, и этот процесс повторяется 5 раз).
scoring: задает метрики для вычисления:
neg_mean_squared_error: отрицательная среднеквадратичная ошибка (MSE). Знак отрицательный, так как в sklearn метрики потерь возвращаются как отрицательные.
r2: коэффициент детерминации, который показывает, насколько хорошо модель объясняет вариацию целевой переменной.
neg_mean_absolute_error: отрицательная средняя абсолютная ошибка (MAE).
return_train_score=True: возвращает оценки на обучающих данных (можно использовать для анализа переобучения).

```python
from sklearn.model_selection import cross_validate
from sklearn.metrics import make_scorer, mean_absolute_error, r2_score
import numpy as np

# Настроим модель GradientBoostingRegressor
gbr = GradientBoostingRegressor(
    n_estimators=1000,
    learning_rate=0.1,
    max_depth=5,
    min_samples_split=5,
    min_samples_leaf=2,
    random_state=0
)

# Функция для MAE
mae_scorer = make_scorer(mean_absolute_error)

# Выполнение кросс-валидации с несколькими метриками
cv_results = cross_validate(
    gbr, X_train, y_train, 
    cv=5,  # 5-кратная кросс-валидация
    scoring={'neg_mean_squared_error': 'neg_mean_squared_error',   # Для MSE
             'r2': 'r2',  # Для R²
             'neg_mean_absolute_error': mae_scorer},  # Для MAE
    return_train_score=True
)

# Извлечение метрик
mse_scores = -cv_results['test_neg_mean_squared_error']  # Отрицательное значение MSE
rmse_scores = np.sqrt(mse_scores)
r2_scores = cv_results['test_r2']
mae_scores = -cv_results['test_neg_mean_absolute_error']  # Отрицательное значение MAE

# Вывод статистики
print(f"Среднее MSE: {mse_scores.mean():.4f}")
print(f"Средний RMSE: {rmse_scores.mean():.4f}")
print(f"Среднее R²: {r2_scores.mean():.4f}")
print(f"Среднее MAE: {mae_scores.mean():.4f}")

print(f"Стандартное отклонение MSE: {mse_scores.std():.4f}")
print(f"Стандартное отклонение RMSE: {rmse_scores.std():.4f}")
print(f"Стандартное отклонение R²: {r2_scores.std():.4f}")
print(f"Стандартное отклонение MAE: {mae_scores.std():.4f}")
```
